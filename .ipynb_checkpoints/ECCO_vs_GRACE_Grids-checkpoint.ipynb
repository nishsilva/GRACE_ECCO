Time series comparison on ECCO and GRACE grids:\

1.	rms(GRACE) on GRACE grids
1.  rms(GRACE) on ECCO grids
1.  rms(ECCOv4r5) on GRACE grids
1.  rms(ECCOv4r5) on ECCO grids
1.  rms(GRACE-ECCOv4r5) on GRACE grids
1.  rms(GRACE-ECCOv4r5) on ECCO grids


```python
import gc
import os
import sys
import xarray as xr

import warnings
warnings.filterwarnings("ignore")

from ipynb.fs.full import RMS
from ipynb.fs.full import Utils
import matplotlib.pyplot as plt
from ipynb.fs.full import Filter

# Used to reload another notebook if code was updated
from importlib import reload

from config import input_dir, output_dir
```

```python
# import ecco_v4_py as ecco
import matplotlib.pyplot as plt
```

```python
# Get datasets that have global and temporal means removed

grace_at_grace = xr.open_dataarray(os.path.join(output_dir, 'pb_grace.nc'))

ecco_at_grace = xr.open_dataarray(os.path.join(output_dir, 'ecco-at-grace-rm-means-grace-grid.nc'))
#ecco_at_grace = ecco_at_grace.rename({'__xarray_dataarray_variable__':'pb'})

[grace_grace, ecco_grace, diff] = Utils.get_aligned_data(grace_at_grace, ecco_at_grace)

# Get original data at ECCO resolution
grace_on_ecco = xr.open_dataarray(os.path.join(input_dir, 'aligned_grace.nc'))
ecco_on_ecco = xr.open_dataarray(os.path.join(input_dir, 'aligned_ecco.nc'))
ecco_weight = xr.open_dataarray(os.path.join(input_dir, 'ecco_early_weight.nc'))

grace_on_ecco = grace_on_ecco.where(grace_on_ecco != 0)
ecco_on_ecco = ecco_on_ecco.where(ecco_on_ecco != 0)
```

```python
# Get the weights for each ocean cell in a GRACE grid
weight=xr.open_dataarray(os.path.join(output_dir, 'grace_weight.nc'))
```

```python
# remove time-mean mean
grace_grace = grace_grace - grace_grace.mean(dim='time')
ecco_grace = ecco_grace - ecco_grace.mean(dim='time')
grace_on_ecco = grace_on_ecco - grace_on_ecco.mean(dim='time')
ecco_on_ecco = ecco_on_ecco - ecco_on_ecco.mean(dim='time')
```

```python
# test without weight
temp = grace_grace.data.reshape((209,360*720))
rms_grace_grace_time = np.nanstd(temp,axis=-1)
temp = grace_on_ecco.data.reshape((209,13*90*90))
rms_grace_on_ecco_time = np.nanstd(temp,axis=-1)

temp = ecco_grace.data.reshape((209,360*720))
rms_ecco_grace_time = np.nanstd(temp,axis=-1)
temp = ecco_on_ecco.data.reshape((209,13*90*90))
rms_ecco_on_ecco_time = np.nanstd(temp,axis=-1)

temp = (grace_grace.data-ecco_grace.data).reshape((209,360*720))
rms_diff_grace_time = np.nanstd(temp,axis=-1)
temp = (grace_on_ecco.data-ecco_on_ecco.data).reshape((209,13*90*90))
rms_diff_on_ecco_time = np.nanstd(temp,axis=-1)
```

```python
plt.figure(figsize=(10,4))
plt.plot(grace_grace.time,rms_grace_grace_time,label='GRACE at GRACE',color='b',linewidth=1)
plt.plot(grace_grace.time,rms_grace_on_ecco_time,'--',label='GRACE at ECCO',color='b',linewidth=1)

plt.plot(grace_grace.time,rms_ecco_grace_time,label='ECCO at GRACE',color='r',linewidth=1)
plt.plot(grace_grace.time,rms_ecco_on_ecco_time,'--',label='ECCO at ECCO',color='r',linewidth=1)

plt.plot(grace_grace.time,rms_diff_grace_time,label='GRACE-ECCO at GRACE',color='g',linewidth=1)
plt.plot(grace_grace.time,rms_diff_on_ecco_time,'--',label='GRACE-ECCO at ECCO',color='g',linewidth=1)

plt.title('std, GRACE vs ECCO')
plt.grid()
plt.legend(ncols=3)
```
