#### Explanation on What happens in this code
Basically, this code processes GRACE data to be analyzed and to be aligned with ECCO data. Below are the exact steps taken.

    1. First reads in GRACE data.
    2. Removes the land values by using a mask
    3. Remove certain dates in the data to avoid making duplicate values when aligned with ECCO times.
    4 Compute ocean spatial mean using a weighted approach and remove ocean spatial mean at each time-setp.
    5. Then subtract this spatial mean at each timestep (gives a spatial mean removed data set).
    6. Create a time dimension for GRACE as a year decimal system
        * time = year + (months-1)/12
    7.  Time mean removed from GRACE OBP.
    8. GRACE OBP saved as a .nc file (where OBP has no time or spatial mean)
    9. Read in above GRACE file.
    10. Define a funtion that regrids GRACE to ECCO
    11. Run the function on GRACE data
    12. Save the data as GRACE_aligned (name varies depending on which version of GRACE) .nc file

```python
# I think remove the data for days that are not day>=12 and day<=19
# before computing spatial mean
da_ocn = da_ocn.sel(time=(da_ocn['time'].dt.day >= 12) & (da_ocn['time'].dt.day <= 19))
```

```python
# remove ocean spatial mean at each time step
R = 6371  # Earth radius, km

area_lat = R*np.deg2rad(0.5)*R*np.cos(np.deg2rad(da_ocn.lat))*np.deg2rad(0.5)
area_lat = area_lat.values
area_latlon = np.tile(area_lat,(len(da_ocn.lon),1)).transpose()  # 2D array (360, 720)

# area of ocean in each cell, of 360x720 cells
area_ocn = area_latlon*(1-mask['land_mask'].values)
area_ocn[area_ocn==0] = np.nan
del R, area_lat, area_latlon

# Global mean of water thickness? 
# sum (water thickness * area_ocn) / area_ocn
# This is an array of 215 monthly global ocean water thickness means.  (Or pressure means?)
glob_mean_GRACE = np.sum(np.sum(da_ocn*area_ocn,axis=1),axis=1)/np.nansum(area_ocn)
# Values are like -1.22852505, -1.79169189, 3.85472235,  3.79361567
np.set_printoptions(suppress=True)

# Subtract the water thickness mean from each da_ocn cell
# (time: 215, lat: 360, lon: 720)
da_ocn_nospatial = da_ocn - glob_mean_GRACE
```

```python
pb_grace = da_ocn_nospatial.values # time*lat*lon
lat_grace = da_ocn_nospatial.lat.values
lon_grace = da_ocn_nospatial.lon.values
years = da_ocn_nospatial.time.values.astype('datetime64[Y]').astype(int) + 1970
months = da_ocn_nospatial.time.values.astype('datetime64[M]').astype(int) % 12 + 1

# "day" in "time" coordinate represent the mid-point of days in a month used to generate monthly mean values.
# Another thing to note is the day in the date variable. When we will only use the data when day>=12 and day<=19. 
# This is not included in the attached codes so feel free to add it.
dates = da_ocn_nospatial['time'].values.astype('datetime64[D]')
days = dates - dates.astype('datetime64[M]') + 1
days = days.astype(int)
del dates

#del grace_filename, mask_filename, da_ocn_nospatial, da_ocn
gc.collect();
```

```python
def grace_to_ecco(grace_grid, method='bilinear'):

    ecco_grid = xr.open_dataset(os.path.join(input_dir_ecco_grid, 'ECCO-GRID.nc'))

    # Extract XG and YG from ecco_grid (3D arrays: tile, j_g, i_g)
    XG = ecco_grid['XG']  # Longitude-like coordinates for the curvilinear grid
    YG = ecco_grid['YG']  # Latitude-like coordinates for the curvilinear grid

    # Get the source lat-lon DataArray
    grace_lats = grace_grid.coords['lat']  # Latitude (shape: 360)
    grace_lons = grace_grid.coords['lon']  # Longitude (shape: 720)

    # Try making nans zero and see what happens
    # grace_grid = grace_grid.fillna(0).  Doesn't make a diference

    # list to store regridded results for each tile
    regridded_tiles = []

    # Regrid each tile's 2D curvilinear grid
    for tile in range(XG.shape[0]):  # Loop over each tile (ecco_pb has 13 tiles)
        # Extract the 2D lat-lon coordinates for the current tile
        XG_tile = XG.isel(tile=tile)  # Shape: (j_g, i_g)
        YG_tile = YG.isel(tile=tile)  # Shape: (j_g, i_g)
        
        # Create a dataset for the current tile's curvilinear grid
        ecco_tile_ds = xr.Dataset({'lon': XG_tile, 'lat': YG_tile})

        # Create the regridder for the current tile
        regridder = xe.Regridder(grace_grid, ecco_tile_ds, method=method)

        # Perform the regridding for the current tile
        regridded_tile = regridder(grace_grid)

        # Append the regridded data for this tile to the list
        regridded_tiles.append(regridded_tile)

    # Combine the regridded tiles back into a single DataArray
    grace_to_ecco_grid = xr.concat(regridded_tiles, dim='tile')
    grace_to_ecco_grid.attrs['units'] = ecco_pb.units

    return grace_to_ecco_grid
```

```python

# Convert grace data to ECCO grid
grace_to_ecco_grid = grace_to_ecco(grace_pb['pb'])

```

```python
print(grace_to_ecco_grid)
grace_to_ecco_grid = grace_to_ecco_grid.rename({'j_g': 'j', 'i_g': 'i'})
```

```python
# write grace_to_ecco_grid from xarray dataarray to xarray dataset
grace_to_ecco_grid = grace_to_ecco_grid.to_dataset(name='pb')

# Save the regridded data to a NetCDF file
#grace_to_ecco_grid.to_netcdf(os.path.join(input_dir, 'grace_regridded_to_ecco.nc'))
```
