## Label GRACE data to define regions of points that go together.

```python
# Load the NetCDF file
grace_pb = xr.open_dataset(os.path.join(input_dir, 'pb_grace.nc'))

# Extract latitude, longitude, and data values
lat = grace_pb['lat'][:]
lon = grace_pb['lon'][:]
print(grace_pb['pb'].shape)
data = grace_pb['pb'][2,:,:]

# Create a mask for the ocean (exclude land and NaN)
ocean_mask = (data != 0) & ~np.isnan(data)
ocean_data = data.where(ocean_mask)

# Ignore zeros and NaNs
unique_values = np.unique(ocean_data)
print(f'number of cells: {len(unique_values)}')

# Prepare an array for labeled regions
labeled_array = np.zeros_like(data, dtype=int)
region_counter = 0

# Label regions for each unique value
for value in unique_values:
    mask = ocean_data == value  # Mask for the current value
    labels, num_features = label(mask)  # Label contiguous regions
    labeled_array += (labels > 0) * (labels + region_counter)  # Offset labels for uniqueness
    region_counter += num_features

# Visualize the labeled array
plt.figure(figsize=(24,16), dpi= 90);
plt.imshow(labeled_array, origin='lower',  interpolation='none')  #cmap='tab20',
plt.colorbar(label='Region Label', orientation='horizontal')
plt.title('GRACE Regions Grouped by Data Value')
# plt.savefig(os.path.join(input_dir, 'GRACE-data-grouped-into-cells.png'), format='png', dpi=90)
plt.show()



#number of cells: 3597
```

```python

def ecco_to_grace(ecco_data):
    '''
    Convert ECCO tiled data to GRACE lat-lon, at the exact GRACE resolution

    '''
    new_grid_delta_lat = .5
    new_grid_delta_lon = .5

    new_grid_min_lat = -90
    new_grid_max_lat = 90

    new_grid_min_lon = -180
    new_grid_max_lon = 180

    # When I didn't have this line, the data looked like GRACE and trended up
    #ecco_data = ecco_data.where(ecco_data!=0)
    ecco_data = np.where(ecco_data == 0, np.nan, ecco_data)

    plot_latlon =\
            ecco.resample_to_latlon(ecco_grid.XC, \
                                    ecco_grid.YC, \
                                    ecco_data,\
                                    new_grid_min_lat, new_grid_max_lat, new_grid_delta_lat,\
                                    new_grid_min_lon, new_grid_max_lon, new_grid_delta_lon,\
                                    fill_value = np.nan,\
                                    mapping_method = 'nearest_neighbor',
                                    radius_of_influence = 120000)


    xr.DataArray(plot_latlon[4]).to_netcdf(os.path.join(input_dir, 'plot_latlon_ecco_data.nc'))

    new_data = plot_latlon[4]

    # Ensure group labels are integers
    group_labels = np.unique(labeled_array)  #[labeled_array > 0]) 

    # Create an array to store averaged data
    averaged_data_grid = np.zeros_like(new_data)

    # Calculate the average for each group
    for lab in group_labels:
   
        # Mask points belonging to the current group
        mask = labeled_array == lab
        
        # Average the data values for the current group
        group_average = np.nanmean(new_data[mask])  # boolean index did not match indexed array along dimension 0; dimension is 375 but corresponding boolean dimension is 360

        # Assign the group average to the corresponding points in the output grid
        averaged_data_grid[mask] = group_average



    averaged_data_grid[averaged_data_grid == 0] = np.nan

    # TODO - I really need a mask here instead of finding the value that land somehow gets.
    #averaged_data_grid[np.isclose(averaged_data_grid, -0.47737114, atol=0.0000001)] = np.nan
    #averaged_data_grid[np.isclose(averaged_data_grid, -2.39774299, atol=0.0000001)] = np.nan

    # find the value that is in the data the most.  All of the land gets the same near zero, but not exactly zero value.
    # I could still use a mask, but I need to get the dimensions and resolution lined up exactly.
    mode = stats.mode(averaged_data_grid).mode[0]
    averaged_data_grid[np.isclose(averaged_data_grid, mode, atol=0.0000001)] = np.nan

    # Save the result - only if testing on one time step
    #xr.DataArray(averaged_data_grid).to_netcdf(os.path.join(input_dir, f'{file_name}.nc'))

    return averaged_data_grid
```

```python
# Apply ecco_to_grace to evey time slice in dataset.
def all_ecco_to_grace(data, name):
 
    
    data = data.where(data != 0, np.nan)
    #data = np.where(data == 0, np.nan, data)
    
    grace_res = xr.apply_ufunc(
                ecco_to_grace,                          # The function to apply
                data,                                    # The DataArray (data to fit)
                input_core_dims=[['tile', 'i', 'j']],    # Specify that the function works along 'time'
                output_core_dims=[['y', 'x']], 
                vectorize=True,                          # Apply to each (i, tile, j) combination
                dask='allowed',                         # Parallelize with dask if needed
                output_dtypes=[float],                   # Output data type is float
                )
    


    # Format dimensions to look like GRACE data
    lat_grace = np.arange(-89.75, 90.0, 0.5)
    lon_grace = np.arange(-179.75, 180.0, 0.5)
    grace_res = Utils.convert_time(grace_res, 1992)
    ds = xr.Dataset(
        data_vars=dict(
            pb=(['time','y', 'x'], grace_res.data, {'units': 'cm/month'}),
        ),
        coords=dict(
            lon=(['x'], lon_grace),
            lat=(['y'], lat_grace),
            time=(['time'],grace_res['time'].data)
        ),
        attrs=dict(description='ECCO at GRACE resolution'),
    )
    ds.to_netcdf(os.path.join(input_dir, f'{name}.nc'))
       
    return grace_res
```

```python
def remove_global_mean(da_ocn):
    '''
    Remove global mean from data on GRACE Grid.

    '''
    grace_weights = xr.open_dataarray(os.path.join(output_dir, 'grace_weight.nc'))
    grace_weights = grace_weights.squeeze()
  
    if 'i' in da_ocn.dims or 'i' in da_ocn.coords:
        da_ocn = da_ocn.rename({'j': 'y', 'i': 'x'})
        #da_ocn = da_ocn.set_index(j='y', i='x')

    if 'lat' in da_ocn.dims or 'lat' in da_ocn.coords:
        da_ocn = da_ocn.rename({'lat': 'y', 'lon': 'x'})
        #da_ocn = da_ocn.set_index(y='y', x='x')

    #grace_weights = grace_weights.assign_coords(x=da_ocn['x'], y=da_ocn['y'])

    #grace_weights = grace_weights.expand_dims(dim='time', axis=0)

    da_ocn = da_ocn.astype('float64')
    #da_ocn = da_ocn.assign_coords(x=np.arange(da_ocn.sizes['x']), y=np.arange(da_ocn.sizes['y']))
    #grace_weights = grace_weights.assign_coords(x=np.arange(grace_weights.sizes['x']), y=np.arange(grace_weights.sizes['y']))
    grace_weights = grace_weights.astype('float64')

    print('sum of grace weights should be 1:')
    print(grace_weights.sum().values)

    # print('grace_weights')
    # print(grace_weights)
    # print('da_ocn')
    # print(da_ocn)

    grace_weights = grace_weights.broadcast_like(da_ocn)
    #grace_weights = grace_weights.set_index(y='y', x='x')
    if (da_ocn['x'].equals(grace_weights['x']) and da_ocn['y'].equals(grace_weights['y'])):
        print('x and y coordinates match!')
    else:
        print('Coordinates mismatch!')

    da_ocn = da_ocn.squeeze()


    print()
    print(da_ocn.dims, da_ocn.sizes)  # ('time', 'y', 'x') Frozen({'time': 375, 'y': 360, 'x': 720})
    print(grace_weights.dims, grace_weights.sizes)   #('time', 'y', 'x', 'j', 'i') Frozen({'time': 375, 'y': 360, 'x': 720, 'j': 360, 'i': 720})
    print()

    glob_mean = (da_ocn * grace_weights).sum(dim=['y', 'x'])
    glob_mean.to_netcdf(os.path.join(output_dir, 'global-mean-of-ecco-to-grace-data.nc'))

    # Subtract the water thickness mean from each da_ocn cell
    global_mean_broadcast = glob_mean.broadcast_like(da_ocn)
    da_ocn_nospatial = da_ocn - global_mean_broadcast

    # da_ocn_nospatial.to_netcdf(os.path.join(output_dir, 'ecco-at-grace-global-mean-rmved.nc'))

    return da_ocn_nospatial
```

```python
def remove_temporal_mean(ds, weight=None):
    '''
    Remove temporal mean first.  That was removed from GRACE data before we ever got it.

    '''

    # NEW:  Remove the temporal mean from each cell.
    # Get time_mean
    # get rid of the 1D variable, if it exists
    ds = ds.squeeze()
    time_mean = np.nanmean(ds, axis=0)

    return ds-time_mean
```
